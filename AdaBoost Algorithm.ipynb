```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('stud.csv')
df
```

```python
print("Categories in 'gender' variable:     ",end=" " )
print(df['gender'].unique())

print("Categories in 'race_ethnicity' variable:  ",end=" ")
print(df['race_ethnicity'].unique())

print("Categories in'parental level of education' variable:",end=" " )
print(df['parental_level_of_education'].unique())

print("Categories in 'lunch' variable:     ",end=" " )
print(df['lunch'].unique())

print("Categories in 'test preparation course' variable:     ",end=" " )
print(df['test_preparation_course'].unique())
```

```python
df = pd.get_dummies(df, columns=["gender","race_ethnicity","parental_level_of_education","lunch","test_preparation_course"])
```

```python
df.head(5)
```

```python
df
```

```python
df['total score'] = df['math_score'] + df['reading_score'] + df['writing_score']
df['percentage'] = df['total score']/3
df.head()
```

```python
df.drop(columns=["math_score","reading_score","writing_score","total score"], inplace=True)	
df.head(3)
```

```python
df = df.astype(int)
df.head(3)
```

```python
df.shape
```

```python
df.drop(columns=["gender_female","race_ethnicity_group E","parental_level_of_education_some high school","lunch_standard","test_preparation_course_none"], inplace=True)	
df.head(3)
```

```python
df.shape
```

```python
X = df.drop(columns=["percentage"])
Y = df["percentage"]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 0)
```

```python
X_train.shape, X_test.shape
```

```python
y_train.shape, y_test.shape
```

```python
from sklearn.ensemble import RandomForestRegressor,AdaBoostRegressor
model = AdaBoostRegressor(n_estimators=50, learning_rate=1, random_state=0)
```

```python
model.fit(X=X_train, y=y_train)
```

```python
y_predicted=model.predict(X_test)
y_predicted
```

```python
y_test
```

```python
from sklearn.metrics import mean_squared_error 
print("mean square error is", mean_squared_error(y_test, y_predicted))
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_predicted)
print("Mean Absolute Error:", mae)
```

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test,  y_predicted))
print("Root Mean Squared Error:", rmse)
```

```python
from sklearn.metrics import r2_score
r2_score1=r2_score(y_test,y_predicted)
print(r2_score1)
```
